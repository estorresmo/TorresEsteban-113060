```python
#ejercicio5

'''defino una funcion contador mediante def, luego mediante un bucle while
comienzo a iterar el numero de la funcion con el fin de encontrar los restos
del valor anteriormente nombrado'''

def cont_5(x):
    cont=0
    while x>0:
        dig1=x%10
        x=x//10
        dig2=x%10
        if dig1==5 and dig2==5:
            cont+=2
            num1=x//10
            dig3=num1%10
            if dig2==5 and dig3==5:
                cont+=1
                if dig1==5 and dig3==5:
                    cont-=2
    return cont

'''en el siguiente ciclo while generamos la variable de entrada para ser 
ingresada por el usuario a forma de entero mediante int+input, luego con 
un ciclo if evaluo la variable de entrada con el finde conocer si se 
ingreso correctamente'''

while True:
    try:
        num=int(input(f"escriba un número entero con varios cincos: "))
        if num<0:
            num*=(-1)
        break
    except ValueError:
        print("el caracter ingresado es incorrecto. Inténtelo de nuevo.")

'''finalmente creo un contador evaluando la funcion contadora inicial con
el fin de poder imprimir el resultado correcto del programa'''

cont=cont_5(num)
if cont>0:
    print("el número de cincos consecutivos que contiene el número\
 ingresado es",cont,".")
else:
    print("el número ingresado NO contiene cincos consecutivos.")
```

```python
#ejercicio6


''' defino una funcion que en este caso sera una suma de fibonacci creando
tambien una lista vacia de la misma, luego defino variables como lo son a,b 
y se igualaran a 1 para asi ralizar la suma de fibonacci'''

def fib(n):                            
    fibonacci=[]
    a,b=1,1
    while a<=n:
        fibonacci.append(a)
        a,b=b,b+a
    return fibonacci

'''defino la variable de entrada para ser ingresada por el usuario
mediante int + input para ser usado como entero y un valor de suma 
igual a cero como parametro inicial'''

num=int(input('ingrese un número: '))
suma=0

'''con el while creo un bucle con el fin de q se itere varias veces
la operacion mostrada a continuacion haciendo restos de cada operacion 
y sumandolos a la variable suma inicial'''

while num/10>0:
    n=num%10
    num=int(num/10)
    suma=suma+n

'''con este ultimo if evaluo la variable suma en la funcion fib para asi
saber si su suma es un numero fibonacci e imprimirlo'''
    
if suma in fib(suma):
    print('la suma de los dígitos del número ingresado es un número de \
          fibonacci.')
else:
    print('la suma del número ingresado no es un número de fibonnaci.')
```

```python
#ejercicio8

''' defino variables para q hacer mas sencilla la ejecucion de operaciones
mas adelante'''

rang_inf=1
rang_sup=101

lim_3=16
ini_4=lim_3-5

mult_3= [3]
mult_4=[]

'''se define la funcion multiplos con el fin de comenzar las operaciones
para conocer el resto ron respecto a el multiplo deseado y comprar este 
resto con 0 '''

def multiplos(valor,multiplo):
    
    resto=valor%multiplo
    
    if resto==0:
        return True
    else:
        return False

''' mediante este ciclo determino los valores multiplos de 3 en los 
primeros 15 valores encontrados del mismo y los adjunto a la lista creada 
en un principio y la imprimo'''
    
for i in range (rang_inf,rang_sup):
    if multiplos(i,3):
        mult_3.append(i)
print('los primeros 15 multiplos de 3 son: ',mult_3[rang_inf:lim_3])

''' mediante este ciclo determino los valores multiplos de 4 en los 
valores restantes del rango, es decir, los multiplos de 4 para los
valores depues el ultimo multiplo de 3 y los adjunto a la lista creada 
en un principio y la imprimo'''

for j in range (rang_inf,rang_sup):
    if multiplos(j,4):
        mult_4.append(j)

print('los multiplos de 4 de restantes son: ',mult_4[ini_4:rang_sup])
```

```python
#ejercicio9

'''defino una funcion por la cual seran evaluadas las coordenadas de dos
circulos independientes entre si y definimos la formulacion para la 
circunferencia y el radio de estas figuras'''

def pun_d2_cir(x1,y1,r1,x2,y2,r2,a,b):

    per_c1=(a-x1)**2+(b-y1)**2 ; per_c2=(a-x2)**2+(b-y2)**2
    rad1=r1**2                      ; rad2=r2**2
    
    return per_c1,rad1,rad2,per_c2

'''dentro del bucle definimos las variables q podran ser ingresadas por 
el usuario como posicion x,y y radio de ambos circulos, de no ser elemtos
numericos el programa no se ejecutara'''

while True:
    try:
        x1=float(input("ingrese la coordenada X del primer circulo = "))
        y1=float(input("ingrese la coordenada Y del primer circulo = "))
        r1=float(input("ingrese el radio del primer circulo = "))
        x2=float(input("ingrese la coordenada X del segundo circulo = "))
        y2=float(input("ingrese la coordenada Y del segundo circulo = "))
        r2=float(input("ingrese el radio del segundo circulo = "))
        a=float(input("ingrese la coordenada X a calcular = "))
        b=float(input("ingrese la coordenada Y a calcular = "))
        break
    except ValueError:
        print("el caracter ingresado es incorrecto. Inténtelo de nuevo.")


'''definimos una igualdad para referenciar la formulacion contra la funcion
definida inicialmente y luego mediante un ciclo if comparar dicha formulacion
y asi poder imprimir lo deseado'''

per_c1,rad1,rad2,per_c2=pun_d2_cir(x1,y1,r1,x2,y2,r2,a,b)
if per_c1<=rad1 and per_c1<=rad2:
    print("el punto pertenece a las dos circulos.")
elif per_c1<=rad1:
    print("el punto pertenece al circulo 1.")
elif per_c2<=rad2:
    print("el punto pertenece al circulo 2.")
else:
    print("el punto no pertenece a ningun circulo.")
```

```python
#ejercicio11

'''importo la libreria keywor con el fin de tener las palabras claves 
discriminadas dentro del codigo y asi poder identificarlas con el
programa ejecutable'''

import keyword  
cad=input('ingrese una cadena de texto: ')

'''defino una varible contador y asigno una lista con vocales mayusculas 
y vocales mayusculas con tilde para q mediante un ciclo for poder 
identificarlas y simplemente sumar la cantidad de estas dentro del contador 
e imprimir'''

cont=0
voc_may=['A','E','I','O','U','Á','É','Í','Ó','Ú','Ü']

for x in cad:
    if x in voc_may and cad:
        cont+=1
print('la cantidad de vocales en mayusculas son: ',cont)

'''defino otra varible contador y asigno una lista con minusculas 
y mayusculas con tilde para q mediante un ciclo for poder identificarlas
y simplemente sumar la cantidad de estas dentro del contador e imprimir'''

cont_2 = 0
voc_til = ['á','é','í','ó','ú','Á','É','Í','Ó','Ú']

for x in cad:
    if x in voc_til and cad:
        cont_2+=1
print('la cantidad de tildes son: ',cont_2)

'''defino otra varible contador y mediante un ciclo for y el uso de 
.isdigit() encontrar la ocurrencia de digitos presentes en el texto
y simplemente sumar la cantidad de estas dentro del contador e imprimir'''

cont_3=0
for x in cad:
    if x.isdigit():
        cont_3+=1
print('la cantidad de dígitos que tiene la cadena son: ', cont_3)

'''defino una varible contador y asigno una lista con (' ') que simboliza
el espacio dentro de un texto q mediante un ciclo for podemos identificarlos
y simplemente sumar la cantidad de estos dentro del contador e imprimir'''

cont_4=0
esp=[' ']

for x in cad:
    if x in esp and cad:
        cont_4+=1
print('la cantidad de espacios son: ', cont_4)

'''defino una varible contador, una de palabras reservadas, y otra
de cadena con .split() con el fin de que me discrimine los elementos
dentro de un texto q mediante un ciclo for podemos identificarlos
y simplemente sumar la cantidad de estos dentro del contador e imprimir'''

cont_5=0
reser=keyword.kwlist
cade=cad.split()

for x in cade:
    if x in reser and cade:
        cont_5+=1
print('la cantidad de palabras reservadas son: ',cont_5)
```

```python
#ejercicio13

'''se lee una cadena de texto q sera ingresada por el usurio mediante 
input'''

cad=input('Digite la cadena de texto: ')

'''se remplazan los espacios dentro del texto por posciciones sin espacios y
se determina el numero de caracteres dentro de del mismo'''

cad=cad.replace(' ', '')
num_car=len(cad)

'''se determinan el # de filas y caracteres maximo por fila que mediante 
un ciclo while seran iteradas con el fin de encontrar dicho valor'''

temp=0    
num_fil=1                  
while temp<num_car:
    temp+=2*num_fil-1
    num_fil+= 1

num_max=2*num_fil-1   

'''a traves de las siguientes variables almacenamos poscicion de cada fila
la cantidad de caracteres y un contador que me verifica si existen letras
aun sin imprimir, luego mediante los ciclos for generar el triangulo e
imprimirlo como se pide inicialmente'''

aux_1=0       
aux_2=0        
cont=0     

for i in range(1,num_fil):
    aux_2 += 2*i-1     
    print(cad[aux_1:aux_2].center(num_max,' '))
    aux_1 = aux_2      
    cont+=1
    
if cont !=num_car:
    print(cad[aux_2:].center(num_max,' '))
```
